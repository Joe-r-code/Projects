```python
ftp_path = 'ftp://ftp.ncbi.nlm.nih.gov/genomes/all/'
```

```python
accessions = []

accessions = list(ncbi_cannabis_genomes['Assembly Accession'])
assembly_name = list(ncbi_cannabis_genomes['Assembly Name'])
```

```python
accessions
assembly_name
```

```python
gca_accessions = [i for i in accessions if 'GCA' in i]
gcf_accessions = [i for i in accessions if 'GCF' in i]
gca_accessions
gcf_accessions
```

```python
for i in gcf_accessions:
    f1 = i[0:3]
    f2 = i[4:7]
    f3 = i[7:10]
    f4 = i[10:13]
    f5 = i
    for j in assembly_name:
        genome_path = (f'{ftp_path}{f1}/{f2}/{f3}/{f4}/{f5}_{assembly_name[0]}/{f5}_{assembly_name[0]}_genomic.fna.gz')
print(genome_path)
```

```python
filename = wget.download(genome_path)
```
